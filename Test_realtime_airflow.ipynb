```python
import sqlite3
```

```python
table_time_cols = {'power':'time',
                   'observation':'obs_time',
                   'forecast':'update_time'}
```

```python
sql_db_path = './realtime/realtime_data/realtime.db'
#sql_db_path = './realtime/realtime_data/realtime_old.db'
#sql_db_path = './realtime/realtime_data/realtime_temp.db'
#sql_db_path = './sql_db/realtime.db'
```

```python
def show_sql_data(table):
    print(table)
    cursor.execute(f'SELECT * FROM {table} ORDER BY {table_time_cols[table]} DESC LIMIT 200;')
    queried_rows = cursor.fetchall()
    for row in queried_rows:
        print(row)
```

```python
conn = sqlite3.connect(sql_db_path)
cursor = conn.cursor()


tables = ['power', 'observation', 'forecast']

for table in tables:
    show_sql_data(table)

cursor.close()
conn.close()
```

```python
date_str = '2024-08-25'
date_str_2nd_day = '2024-08-26'

conn = sqlite3.connect(sql_db_path)
cursor = conn.cursor()

sql_command = f"SELECT * FROM power WHERE time > '{date_str} 00:00' AND time <= '{date_str_2nd_day} 00:00' AND generator = '電池'"
#print(sql_command)
cursor.execute(sql_command)

sql_output = cursor.fetchall()
cursor.close()
conn.close()
```
